$\textit{1. Método de Newton}$

```julia
function Newton(f,df,x0,N)
    c=0
    x1=0.0
    while c<N
        x1=x0-(f(x0)/df(x0))
        x0=x1
        c=c+1
    end
    return x1
end
```

```julia
g(x)=x^2-2
dg(x)=2x
```

```julia
Newton(g,dg,1,10)
```

$\textit{2. Error y convergencia}$

```julia
function errorlocal(f,df,x0,N,xreal)
    listax0=[]
    listaN=[]
    
    for i=1:N
        push!(listaN,i)
        Ne=Newton(f,df,x0,N)
        push!(listax0,Ne)
    end
    return(listaN,listax0)
end
```

Error en función del número de pasos.

```julia
xreal=sqrt(2)
EL=errorlocal(g,dg,1,50,xreal)

scatter(EL[1],EL[2], xlabel="x", ylabel="f(x)", title = "Error Newton", color="blue", label="Error", 
    legend = true, linewidth = 1, grid = true)
```

$\textit{Dependencia de la raíz encontrada en términos de la condición inicial para }f(x)=x^2-2 $

Ahora generaremos dos listas listas, una con condiciones iniciales y otra con la raíz que regresa Newton en función de las condiciones iniciales.

```julia
m=[]
m1=[]
for i= -3:0.125:3
    push!(m,Newton(g,dg,i,10))
    push!(m1,i)
end
```

```julia
plot(m1,m, xlabel="x", ylabel="f(x)", title = "Error Newton", color="blue", label="Error", 
    legend = true, linewidth = 1, grid = true)
```
